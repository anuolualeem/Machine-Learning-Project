Our Objectives for this project is:
* To predict patients suffering from heart disease using seven different machine learning classifiers such as K-Nearest Neighbors, SVM, Decision Tree, Random Forest, Logistic Regression, Neural Network, and AdaBoost classifier

* And to estimate and compare the performance of all the classifiers, in order to select the one with the best performance.

## Methods
To achieve these objectives, we will follow the steps below

* Exploratory Data Analysis
* Predictive analysis

### Data Preprocessing
* For this project, we will use the Cleveland Heart Disease dataset downloaded in csv format from UCI Repository link [here](https://archive.ics.uci.edu/ml/datasets/heart+disease).
* Its attributes are described below:

|Attribute name| Type| Descrption
|---------|-------------|------------|
|age |continuous | patient age in years |
|sex   |discrete | patient gender (1=male; 0=female) |
| cp	|discrete  | chest pain type (1 = typical angina; 2 = atypical angina; 3 = nonanginal pain; 4 = no pain)	|
| trestbps	|continuous | resting blood pressure (in mm Hg) on admission to the hospital |
| chol	|continuous| serum cholestoral in mg/dl |
| fbs	|discrete| (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) |
| restecg |discrete| resting electrocardiographic (0 = normal; 1 = ST-T wave abnormality; 2 = probable/definite left ventricular hypertrophy)|
| thalach	|continuous| maximum heart rate achieved |
| exang	|discrete| exercise induced angina (1 = yes; 0 = no) |
| oldpeak	|continuous| ST depression induced by exercise relative to rest |
| slope	|discrete| the slope of the peak exercise ST segment (1 = up-sloping; 2 = flat; 3 = down-sloping)|
| ca	|discrete| number of major vessels (0-3) colored by flourosopy |
| thal	|discrete| thallium heart scan (3 = normal; 6 = fixed defect; 7 = reversible defect) |
| num	|discrete (binary)| diagnosis of heart disease (angiographic disease status) (0 = absent; 1 to 4 = present)|

Now let's get started by loading the data into python

```python
# Import needed libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# sklearn libraries for machine learning algorithms
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

# Load dataset
colnames=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
         'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
disease = pd.read_csv('processed.cleveland.csv', names=colnames, na_values='?')
print(disease.shape)
disease.head()
```

```python
# check the data types of columns
disease.dtypes
```

```python
# change the datatype of the discrete features from string/object to category
cat_columns = ["sex", "cp", "fbs", "restecg", "exang", "slope", "ca", "thal", "num"]
int_columns = ["age", "trestbps", "chol", "thalach"]
for i in disease.columns:
    if i in cat_columns:
        if i == "num":
            # regroup to binary class
            disease[i] = disease[i].map({0 : 0, 1: 1, 2 : 1, 3 : 1, 4 : 1})
        disease[i] = pd.Categorical(disease[i].astype('Int64'))
    elif i in int_columns:
        disease[i] = disease[i].astype('Int64')
        
print(disease.dtypes)
disease.head()
```

```python
# Checking for missing values
disease.info()
```

```python
# drop the missing values
disease = disease.dropna()
disease.info()
```

### Predictive Analysis using Machine learning
Now we will train our dataset using 7 machine learning classifiers: K Nearest Neighbors (KNN), Support Vector Machine (SVM), Decision Tree, Random Forest, Logistic Regression, Neural Network, AdaBoost Classifiers. We will evaluate and compare their performance, and best performing classifier will be selected.

Let's prepare our data for the analysis

```python
X = disease
# Convert each category to numerical data type
for i in cat_columns:
    X[i] = X[i].astype(int)
    
# seperate features and label variables and split into training and test datasets
y = X.iloc[:,-1]
X = X.iloc[:,:-1]

# split dataset into 70% training data and 30% test data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

### K Nearest Neighbors (KNN) Classifier
Let's train KNN on the training data using k = 1,3,5,15,25, and 50 nearest neighbors, and 2 types of weight. We will calculate the average of accuracy, precision, recall and f1 scores, and select the one with the highest average score.

```python
# function that computes classifier performance
def perf(classifier, X_test, y_test, y_pred):
    acc = round(classifier.score(X_test, y_test), 2)
    p = round(precision_score(y_test, y_pred), 2)
    r = round(recall_score(y_test, y_pred),2)
    f1 = round(f1_score(y_test, y_pred),2)
    results = {"accuracy": acc,
              "precision": p,
              "recall": r,
              "f1": f1,
               "average": round((acc + p + r + f1)/4, 2)
              }
    return(results)
```

```python
k = k = [1,3,5,15,25,50]  # number of neighbors used
weight = ['distance', 'uniform']
scores = []
for i in k:
    for j in weight:
        knn_classifier = KNeighborsClassifier(n_neighbors = i , weights = j)
        knn_classifier.fit(X_train, y_train)
        y_pred = knn_classifier.predict(X_test)
        perf_metrics = perf(knn_classifier, X_test, y_test, y_pred)
        scores.append([perf_metrics['average'], perf_metrics['accuracy'], 
                       perf_metrics['precision'], perf_metrics['recall'], perf_metrics['f1'], j, i
                      ])
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1','weight','k']
print(perf_df)
print("\nmaximum performance",max(scores))
```

5 nearest neighbors with weight = "uniform" gave the highest average performance of 67%

### Support Vector Machine (SVM) Classifier
Let's train SVM on the training data using three kernels: {linear, poly, and rbf} and regularization parameters: 0.025, 0.05, 0.5, 1]

```python
kernels = ["linear", "poly", "rbf"]
c = [0.025, 0.05, 0.5, 1, 5]
scores = []
for i in kernels:
    for j in c:
        svm_classifier = SVC(kernel=i, C=j)
        svm_classifier.fit(X_train, y_train)
        y_pred = svm_classifier.predict(X_test)
        perf_metrics = perf(svm_classifier, X_test, y_test, y_pred)
        scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                       perf_metrics['recall'], perf_metrics['f1'], i, j
                      ])
    
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1','kernel', "reg_parameter"]
print(perf_df)
print("\nmaximum performance",max(scores))
```

Linear SVM with 0.05 regularization parameter gave the highest average performance of 89%

### Decision Tree Classifier
We will use 6 maximum depths of trees : 5, 10, 15, 20, 25, 30 in decision tree to train the data

```python
tree_max_depth = [i for i in range(5,35,5)]
scores = []
for i in tree_max_depth:
    decisionTree_classifier = DecisionTreeClassifier(max_depth=i)
    decisionTree_classifier.fit(X_train, y_train)
    y_pred = decisionTree_classifier.predict(X_test)
    perf_metrics = perf(decisionTree_classifier, X_test, y_test, y_pred)
    scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                   perf_metrics['recall'], perf_metrics['f1'], i
                  ])
    
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1','tree_max_depth']
print(perf_df)
print("\nmaximum performance",max(scores))
```

Decision tree with 5 maximum depth of trees gave the highest average performance of 73%

### Random Forest Classifier
Here, we will use different number of trees: 10, 100, 200, 500, maximum depth of tree:5, 10, 15, 20 and number of features: 1-5 that will be used to train the classifier

```python
tree_no = [10, 100, 200, 500]
tree_max_depth = [i for i in range(5,20,5)]
n_features = [1,2]
scores = []
for i in tree_no:
    for j in tree_max_depth:
        for k in n_features:
            RandomForest_classifier = RandomForestClassifier(max_depth=j, n_estimators=i, max_features=k)
            RandomForest_classifier.fit(X_train, y_train)
            y_pred = RandomForest_classifier.predict(X_test)
            perf_metrics = perf(RandomForest_classifier, X_test, y_test, y_pred)
            scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                           perf_metrics['recall'], perf_metrics['f1'], i, j, k
                          ])
    
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1', 'tree_no', 'max_depth', 'max_features']
print("maximum performance",max(scores))
perf_df
```

Random Forest with 500 number of trees, 5 maximum depth of tree and 1 max feature gave the highest average performance of 88%

### Neural Network
We will use (50 ,100) number of hidden layers with activation function: relu & tanh in each hidden layer, initial learning rate: 0.001, 0.01, 0.1 and alpha: 0.1, 1, which is the strength of regularization term

```python
scores = []
h_layers = [50 ,100]
activation = ["relu", "tanh"]
learning_rate = [0.001, 0.01, 0.1]
alpha = [0.1, 1]
for i in h_layers:
    for j in activation:
        for k in learning_rate:
            for l in alpha:
                nn_classifier = MLPClassifier(hidden_layer_sizes=(i,), activation=j, alpha=l, learning_rate_init=k)
                nn_classifier.fit(X_train, y_train)
                y_pred = nn_classifier.predict(X_test)
                perf_metrics = perf(nn_classifier, X_test, y_test, y_pred)
                scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                               perf_metrics['recall'], perf_metrics['f1'], i, j, k, l
                              ])
    
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1',
                   'hidden_layers', 'activation', 'learning_rate', 'alpha']
print("maximum performance",max(scores))
perf_df
```

Neural Network with 50 hidden layers using relu activation function with an initial learning rate of 0.01 and regularization strength of 1 gave the highest average accuracy of 87%

### Adaboost Classifier
Using 3 different maximum number of estimators:20,50,100 and 3 learning rates:0.01,0.1,1 applied to each classifier at each boosting iteration

```python
scores = []
n_estimators = [20, 50, 100]
learning_rate =  [0.01, 0.1, 1]
for i in n_estimators:
    for j in learning_rate:
        ada_classifier = AdaBoostClassifier(n_estimators=i, learning_rate=j)
        ada_classifier.fit(X_train, y_train)
        y_pred = ada_classifier.predict(X_test)
        perf_metrics = perf(ada_classifier, X_test, y_test, y_pred)
        scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                       perf_metrics['recall'], perf_metrics['f1'], i, j
                      ])
        
perf_df = pd.DataFrame(np.array(scores))
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1','max_estimators', "learning_rate"]

print("\nmaximum performance",max(scores))
perf_df
```

AdaBoost classifier with 100 maximum number of estimators, and 0.1 yields the best average score of 87% out of all the hyperparameter used to train the model.

### Logistic Regression
We will use 3 methods: Lasso, Ridge and Elastic Net Logistic Regression which uses l1, l2 and combination of l1 and l2 regularization respectively.

```python
scores = []
penalty = ["l1","l2","elasticnet"]
for i in penalty:
    if (i == "elasticnet"):
        logisticReg_classifier = LogisticRegression(penalty= i, solver = 'saga', max_iter=10000, l1_ratio=0.5)
    else:
        logisticReg_classifier = LogisticRegression(penalty= i, solver = 'liblinear', max_iter=10000)
    logisticReg_classifier.fit(X_train, y_train)
    y_pred = logisticReg_classifier.predict(X_test)
    perf_metrics = perf(logisticReg_classifier, X_test, y_test, y_pred)
    scores.append([perf_metrics['average'], perf_metrics['accuracy'],perf_metrics['precision'], 
                   perf_metrics['recall'], perf_metrics['f1'], i
                  ])
print(scores)        
perf_df = pd.DataFrame(np.array(scores))
#print(perf_df)
perf_df.columns = ['average_score','accuracy', 'precision', 'recall','f1','penalty']

print("\nmaximum performance",max(scores))
perf_df
```

Elastic Net Logistic Regression turned out to give the best average performance of 89% out of all the Logistic Regression methods used.
